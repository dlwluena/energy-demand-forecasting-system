# Energy Demand Forecasting using Machine Learning

This project aims to predict electricity demand using historical energy consumption data and machine learning techniques. The dataset represents real-world power usage patterns, making this problem a time series forecasting task.

## Dataset Description

The dataset contains hourly electricity demand measurements.

- **Datetime**: Timestamp of the observation  
- **PJME_MW**: Energy consumption in megawatts (MW)

This is a time series dataset, meaning it contains temporal patterns such as daily cycles, weekly trends, and seasonal variations.

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("../data/PJME_hourly.csv")

df["Datetime"] = pd.to_datetime(df["Datetime"])
df = df.set_index("Datetime")

df = df.sort_index()

print(df.head())
```

```python
plt.figure(figsize=(15,5))
plt.plot(df.index, df["PJME_MW"])
plt.title("Electricity Demand Over Time")
plt.xlabel("Date")
plt.ylabel("MW")
plt.show()
```

```python
df["hour"] = df.index.hour

hourly_avg = df.groupby("hour")["PJME_MW"].mean()

hourly_avg.plot(kind="bar", title="Average Demand by Hour")
plt.show()
```
